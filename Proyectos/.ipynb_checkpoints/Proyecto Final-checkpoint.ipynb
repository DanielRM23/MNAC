# Problema:

    
    
<li>
    Dado el siguiente siguiente sistema lineal de ecuaciones diferenciales ordinarias (2D): 
<br>    
    
$
\dot{\vec{x}} = \left[\begin{array}{c} \dot{x_1} \\ \dot{x_2} \end{array}\right]= \left[\begin{array}{c} \alpha x_1 + \beta x_2 \\  \gamma x_1 + \delta x_2 \end{array}\right] = \left[\begin{array}{cc} \alpha  & \beta  \\  \gamma & \delta \end{array}\right] \left[\begin{array}{c} x_1 \\ x_2 \end{array}\right],
$

Escribir un programa que grafique el espacio fase y algunas curvas solución (*i.e.* cuatro condiciones inciciales diferentes) alrededor del punto de equilibrio $(0,0)$. 

Las curvas solución fueron obtenidas numéricamente usando el método de Euler para N-dimensiones. 

 
## EJEMPLOS: 

Oscilador Armonico Simple:

$\dot{x}= y $ <br> $\dot{y}= - k x$; $\;\;$  para diferentes valores de $k$

########################
    
Oscilador Armonico Amortiguado:
    
$\dot{x}= y $ <br> $\dot{y}= - k x - \beta y$; $\;\;$  para diferentes valores de $k$ y $\beta$ 
           
########################               
    
$\dot{\vec{x}}(t) = \mathbb{A} \vec{x}(t) $  ; $\;\;$ donde $\mathbb{A} $    
<br>              

$\mathbb{A}
=
\begin{pmatrix} 
a & 0 \\
0  & -1
\end{pmatrix}
$ ; $\;\;$  para diferentes valores de $a$


```python
#Se importan las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt 
import random
#Estas son las librerías para las animaciones
from ipywidgets import *
from IPython.html.widgets import *
#Esta es para el campo de pendientes
from pylab import quiver
#Python simbolico 
import sympy as sym 
from sympy.vector import *
```

## Esto nomas es pa ver que se va a hacer :s

```python
#Se definen las variables a utilizar 
x1, x2, u, u1, u2, alpha, beta, gamma, delta, lamda , t= sym.symbols("x1 x2 u u1 u2 alpha beta gamma delta lamda t")
```

```python
#Esta es la matriz inicial
A = sym.Matrix( [ [alpha, beta],
                 [gamma, delta] ])

#Vector de dos dimensiones
X = sym.Matrix([x1, x2]) 

#Esta función cambia los valores de la matriz dada por A
def cambios_variables(A, a,b,c,d):
    A = A.subs([(alpha,a)])
    A = A.subs([(beta,b)])
    A = A.subs([(gamma,c)])
    A = A.subs([(delta,d)])
    return A

A
```

```python
#Multiplicación de matriz por vector
A*X
```

```python
#Aquí ya le damos valores a los elementos de la matriz
A_new = cambios_variables(A, 0,1,-1,0)
A_new
```

```python
#Así se ve la solución general del sistema de ecuaciones
sol = sym.exp(lamda*t)*u
sol
```

```python
#Valores propios
lamda1, lamda2 = A_new.eigenvals()
lamda2
```

```python
#Vectores propios.
#En este caso sí ocurren errores si se tienen valores propios iguales

u1 = A_new.eigenvects()[0][2][0]
u2 = A_new.eigenvects()[1][2][0]
u1
```

```python
#Esta es la solución general
solucion = sym.exp(lamda1*t)*u1 + sym.exp(lamda2*t)*u2
solucion.simplify()
solucion
```

```python
#Se puede evaluar 
solucion.subs(t,0).evalf()
```

```python
#Graficamos 

tt = np.linspace(0,10,1000)


#Solucion en x
x_coord = sym.lambdify([t], solucion[0])(tt)
#Solucion en y
y_coord = sym.lambdify([t], solucion[1])(tt)

plt.figure(figsize=[10,6])
plt.grid()
#Línea f(x)=0 y x = 0
plt.axhline(0, color='black')
plt.axvline(0, color='black')
#Nombre de los ejes
plt.xlabel("$x_1$", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
plt.ylabel("$x_2$", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
plt.plot(x_coord, y_coord, color="red")
```

## Esta función solo se define para cambiar el color de las gráficas. Se obtienen colores arbitrarios.

```python
def color():
    """
    Funcion que nos regresa una tupla (r,g,b)
    para colores aleatorios
    """
    r = random.random()
    b = random.random()
    g = random.random()
    color = (r, g, b)
    return color
```

## Se define el método de Euler N-dimensional.

```python
#Se define lo siguiente:
#F: Campo vectorial
#x: vector N-dimensional
#x0: vetor N-dimensional de condiciones iniciales
#t0: tiempo inicial
#tf: tiempo final 
#dt: paso
def EulerND(F,x0,t0,tf,dt):
    """
    Método de Euler N-dimensional
    """
    #En esta lista vacía se guardarán los valores del tiempo
    T = [t0]    
    X = []
    #Se está haciendo una lista de listas de N-dimensiones
    #Es decir, un vector de vectores
    #A los vectores de la lista grande se les está agregando los 
    #valores iniciales
    for i in range(len(x0)):
        a = []
        X.append(a)
        X[i].append(x0[i])
    
    t = t0
    x = x0
    #Aquí se implementa el método general
    while t < tf:
        T.append(t0)
        x = x + F(x)*dt
        t = t +  dt
        for i in range(0,len(x0)):
            X[i].append(x[i])
        T.append(t)
    
    return X,T
```

## Se definen las características del problema planteado, en este caso, se optó por usar clases.

```python
class Sistemas():
    
    #Se definen los 4 parámetros que definen a nuestro objeto
    #Recordemos que se puede obtener una matriz.
    #A = (alpha, beta) 
    #    (gamma, delta)
    def __init__(self, alpha,beta,gamma,delta):
        self.alpha = alpha
        self.beta = beta
        self.gamma = gamma
        self.delta = delta
       
    #En esta parte se define la multiplicación de la matriz A por un vector x
    #El vector es de dos dimensiones 
    def arreglo(self,x):
        y = np.array(np.zeros(2))
        y[0] = self.alpha*x[0] + self.beta*x[1]
        y[1] = self.gamma*x[0] + self.delta*x[1]
        return y
    
    #En esta parte se está graficando las soluciones que nos da Euler
    #Como argumentos de la función se necesita: 
    #x0: vector N-dimensional de condiciones iniciales
    #t0: tiempo inicial
    #tf: tiempo final 
    #dt: paso
    def plot_sistema(self,x0,t0,tf,dt):
        
        #Se usa el método de Euler 
        X1 = EulerND(self.arreglo,x0,t0,tf,dt)[0][0]
        X2 = EulerND(self.arreglo,x0,t0,tf,dt)[0][1]
        T = EulerND(self.arreglo,x0,t0,tf,dt)[1]
        #Se grafica
        #plt.figure(figsize=[12,8])
        plt.plot(X1,X2, "-",
                 label = f"alpha: {self.alpha}\nbeta: {self.beta}\ngamma: {self.gamma} \ndelta:{self.delta}",
                color = color() )
        #Esta es la condición inicial, se grafica en forma de un punto
        plt.plot(x0[0],x0[1],'o',color = color(),label=f"Condicion inicial: ({x0[0]}, {x0[1]})")
        #Estos son los ejes coordenados
        plt.axhline(0, color='black')
        plt.axvline(0, color='black')
        #Título
        plt.title("Espacio Fase $x_1$ vs $x_2$", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
        #Nombre de los ejes
        plt.xlabel("$x_1$", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
        plt.ylabel("$x_2$", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
        #Cuadrícula
        plt.grid()
        plt.legend()
    
    #Esta función nos proporciona condiciones iniciales aleatorias de la forma (m,n) 
    #m y n son los números entre los cuales queremos tomar números aleatorios
    def plot_varias_condiciones_random(self,t0,tf,dt, m, n, N):
        condiciones = [ [random.randint(m,n), random.randint(m,n)] for i in range(N) ]
        plt.figure(figsize=[12,8])
        for i, j in condiciones:
            x0 = np.array([i,j])
            self.plot_sistema(x0,t0,tf,dt)

        plt.title("Espacio Fase $x_1$ vs $x_2$, varias condiciones iniciales", 
                  fontdict={'fontname': "Comic Sans MS", "fontsize": 15})

    #Esta función se usa para graficar, pero debemos dar las condiciones iniciales en forma de un arreglo
    def plot_varias_condiciones(self,t0,tf,dt,condiciones):
        plt.figure(figsize=[12,8])
        for i, j in condiciones:
            x0 = np.array([i,j])
            self.plot_sistema(x0,t0,tf,dt)
            
            
    #Esta función grafica el campo de pendientes del sistema
    def campo_pendientes(self,matriz,x0,t0,tf,dt,m,n):
        plt.figure(figsize=[12,8])
        
        for i in np.arange(m,n+.2,0.2):
            for j in np.arange(m,n+.2,0.2):
                x = np.array([i,j])
                y = matriz(x)
                #Normalizo los vectores
                norma = np.sqrt(y[0]**2 + y[1]**2)
                plt.quiver(i,j,y[0]/norma,y[1]/norma, width = 0.003, scale = 30, color = color())
        
        self.plot_sistema(x0,t0,tf,dt)
        plt.axhline(0, color='black')
        plt.axvline(0, color='black')
        #Título
        plt.title("Campo de pendientes", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
        #Nombre de los ejes
        plt.xlabel("$x_1$", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
        plt.ylabel("$x_2$", fontdict={'fontname': "Comic Sans MS", "fontsize": 15})
        plt.xlim(m-.2, n +.2)
```

```python
#Defino mi objeto con los parámetros de la matriz
s1 = Sistemas(alpha=0, beta=1, gamma=-1, delta=-1)
```

```python
#Grafico mi objeto anterior con las condiciones dadas 
s1.plot_sistema(x0=[1,0],t0=0,tf=10,dt=0.001)
```

```python
#Campo de pendientes

#Esta función nos proporciona condiciones iniciales aleatorias de la forma (m,n) 
#m y n son los números entre los cuales queremos graficar el campo de pendientes
matriz = s1.arreglo
s1.campo_pendientes(matriz,x0=[1,0],t0=0,tf=10,dt=0.001,m=-1,n=1)
```

```python
#Grafico varias condicones iniciales del sistema
#Las condiciones iniciales están dadas de forma aleattoria
#Esta función nos proporciona condiciones iniciales aleatorias de la forma (m,n) 
#m y n son los números entre los cuales queremos tomar números aleatorios
matriz = s1.arreglo
s1.plot_varias_condiciones_random(t0=0,tf=10,dt=0.001,m=-2,n=2,N=3)
```

## Otra vez Euler, pero con una pequeñísima modificación

```python
#Se define lo siguiente:
#F: Campo vectorial
#x: vector N-dimensional
#x0: vetor N-dimensional de condiciones iniciales
#t0: tiempo inicial
#tf: tiempo final 
#dt: paso
def EulerND2(F, x0,t0,tf,dt, params):
    """
    Método de Euler N-dimensional
    """
    #En esta lista vacía se guardarán los valores del tiempo
    T = [t0]    
    X = []
    #Se está haciendo una lista de listas de N-dimensiones
    #Es decir, un vector de vectores
    #A los vectores de la lista grande se les está agregando los 
    #valores iniciales
    for i in range(len(x0)):
        a = []
        X.append(a)
        X[i].append(x0[i])
    
    t = t0
    x = x0
    #Aquí se implementa el método general
    while t < tf:
        T.append(t0)
        x = x + F(x, params)*dt
        t = t +  dt
        for i in range(0,len(x0)):
            X[i].append(x[i])
        T.append(t)
    
    return X,T
```
